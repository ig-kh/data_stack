```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_against(df_default, df_opt):
    nrows = len(df_default.drop_duplicates(subset=["resource","stage"]))
    fig, ax = plt.subplots(nrows, 1, figsize=(9,9))

    for i, (((res, stage), dist1), ((_, _), dist2)) in enumerate(zip(df_default.groupby(["resource","stage"]).agg(list).itertuples(), df_opt.groupby(["resource","stage"]).agg(list).itertuples())):
        name = res+" "+stage
        data = {"default": np.asarray(dist1, dtype=np.float32), "optimized": np.asarray(dist2, dtype=np.float32)}
        ax[i].boxplot(data.values(), labels=data.keys())
        ax[i].axhline(np.median(np.asarray(dist1, dtype=np.float32)))
        ax[i].axhline(np.median(np.asarray(dist2, dtype=np.float32)), c="orange")
        ax[i].set_title(name)
```
